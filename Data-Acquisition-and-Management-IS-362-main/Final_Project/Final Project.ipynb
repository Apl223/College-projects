```python
"""
Project proposal:

I'd like to see which countries consume the most alcohol and match those 
with countries with alcoholism consumption disorder. I want to see if higher amounts 
of average alcohol consumption necessarily means that a given country will 
have higher amounts of alcoholism.

"""

import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv')
data.sort_values("beer_servings", ascending=False).head(10)
```

```python
data.sort_values("spirit_servings", ascending=False).head(10)
```

```python
data.sort_values("wine_servings", ascending=False).head(10)
```

```python
data['Consumption'] = data['beer_servings'] + data['spirit_servings'] + data['wine_servings']
data['Consumption'].sort_values(ascending=False).head(10)
```
