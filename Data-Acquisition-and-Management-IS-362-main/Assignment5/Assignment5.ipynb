```python
import pandas as pd
from IPython.display import Markdown as md
airports = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/master/data-raw/airports.csv")
airports.sort_values("lat", ascending=False).head(8)
```

```python
md("The northen most point is Dilant Hopkins Airport.")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tidyverse/nycflights13/master/data-raw/airports.csv')
northeast_airports = df[(df['lat'] > 44) & (df['lat'] <= 50) & (df['lon'] >= -75) & (df['lon'] <= -65)]
northeast_airports.sort_values('lat', ascending=False)
```

```python
md("Eastern most point is Northern Aroostook Regional Airport")
```

```python
speed = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/master/data-raw/weather.csv")
speed[(speed["month"]==2) & (speed["day"]==12) & (speed["year"]==2013)].sort_values("wind_speed", ascending=False)
```
